```python
import re

# Preparation of file (remove \f)

file = open('dicionario_medico.txt')
lines = file.readlines()
file.close()

newlines = []
for i in range(len(lines)-1):
    if re.match('[A-Z]',lines[i-1]) and re.match('\n',lines[i]) and re.match('\f[a-z]',lines[i+1]) and not(re.match('[A-Z]',lines[i+2])): #if description starts then empty line then \f before rest of description -> omit \n -> rest of description moves up to rest of description
        pass
    elif re.match('\n',lines[i]) and re.match('\f[A-Z]',lines[i+1]):   #if empty line then \f before description -> don't add newline -> description moves on line up to the decleration 
        pass
    elif re.match('\n',lines[i]) and re.match('\f[a-z]',lines[i+1]):   #if empty line then \f before decleration
        newlines.append(lines[i])
    elif re.match('\f',lines[i]):   #sub all \f with nothing
        newlines.append(re.sub('\f','',lines[i]))
    else:
        newlines.append(lines[i])



newtext = ''.join(newlines)


newfile = open('new_dicionario_medico.txt','w')
newfile.write(newtext)
newfile.close()

# import lines

file = open('new_dicionario_medico.txt')
lines = file.read()
file.close()

list = re.findall(r'\n\n(.+)((?:\n.+)+)',lines)

# strip lines

new_entries = [(designation,describtion.strip()) for designation, describtion in list]

# write html

html = open('Dicionario Medico.html','w')

header = '''<html>
<head>
<meta charset='utf-8'/>
</head>
<style>
td {font-size:20px;}
</style>
<body style="background-color:black;">
<h1 style="text-align:center;font-size:60px;color:white;"><u>Dicionario Medico</u></h1>
<table style="border:2px solid white;border-collapse:collapse;">
<caption style="color:white;padding-bottom:20px;">Source: https://github.com/lfcc1/plneb-2223/blob/master/data/dicionario_medico.txt</caption>
<th style="border: 1px solid white;font-size:30px;color:white">Designation</th>
<th style="border: 1px solid white;font-size:30px;color:white">Describiton</th>
'''

body=''
for designation, describtion in new_entries:
    body += '<tr style="height:50px">'
    body += '<td style="border: 1px solid white;color:white;padding-left:7px;">' + designation + '</td>'
    body += '<td style="border: 1px solid white;color:white;padding-left:7px;">' + describtion + '</td>'
    body += '</tr>'

footer = '''
</table>
</body>
</html>'''

html.write(header + body + footer)

html.close()
```
